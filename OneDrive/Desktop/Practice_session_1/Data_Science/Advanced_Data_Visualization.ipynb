```python
import gspread
from google.oauth2.service_account import Credentials

# Scopes required for Sheets and Drive
scopes = ["https://www.googleapis.com/auth/spreadsheets",
          "https://www.googleapis.com/auth/drive"]

creds = Credentials.from_service_account_file("service_account.json", scopes=scopes)
gc = gspread.authorize(creds)

# Open the sheet
sh = gc.open("youtube_data")
worksheet = sh.worksheet("Sheet1")
```

```python
from gspread_dataframe import get_as_dataframe
data = get_as_dataframe(worksheet)
#drop any completely any rows or columns
```

```python

import plotly.figure_factory as ff

# Assuming 'data' is your dataframe
correlation = data[['subscribers', 'video views', 'engagement_rate', 'uploads']].corr()

fig_heatmap = ff.create_annotated_heatmap(
    z=correlation.values,
    x=list(correlation.columns),
    y=list(correlation.index),
    colorscale='Viridis'
)

# Adding title to the layout
fig_heatmap.update_layout(
    title="Heatmap: Correlation Between Numeric Variables",
    xaxis_title="Variables",
    yaxis_title="Variables"
)

fig_heatmap.show()
```

```python
import plotly.graph_objects as go
channel_name = "T-Series"
channel_data = data[data['Title'] == channel_name]
engagement_rate = channel_data['engagement_rate'].iloc[0]

fig_gauge = go.Figure(
    go.Indicator(
        mode="gauge+number",
        value=engagement_rate * 100,  # Convert to percentage
        gauge={
            "axis": {"range": [0, 100]},  # Min and Max in percentage
            "bar": {"color": "green"},
            "steps": [
                {"range": [0, 50], "color": "lightgray"},
                {"range": [50, 100], "color": "lightblue"}
            ]
        },
        title={"text": f"Engagement Rate for {channel_name}"}
    )
)

fig_gauge.show()
```

```python
import plotly.graph_objects as go

# Top 5 channels by video views
top_channels = data.nlargest(5, 'video views')
categories = top_channels['Title'].tolist() + ['Total']
video_views = top_channels['video views'].tolist()
video_views.append(sum(video_views))  # Add total

fig_waterfall = go.Figure(
    go.Waterfall(
        x=categories,
        y=video_views,
        connector={"line": {"color": "rgb(63, 63, 63)"}},
        text=[f"{int(val):,}" for val in video_views],
        textposition="outside"
    )
)

fig_waterfall.update_layout(
    title="Waterfall Chart: Video Views Breakdown for Top Channels",
    xaxis_title="Channels",
    yaxis_title="Video Views",
    showlegend=False
)

fig_waterfall.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a subplot layout with 2 rows and 2 columns
# Add custom titles for each subplot
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Histogram of Subscribers",  # Top-left: Histogram
                    "Line Plot: Video Views vs Rank",  # Top-right: Line Plot
                    "Bar Chart: Engagement Rate by Category",  # Bottom-left: Bar Chart
                    "Scatter Plot: Subscribers vs Video Views")  # Bottom-right: Scatter Plot
)

# Add a Histogram to the first subplot (row 1, col 1)
fig.add_trace(
    go.Histogram(x=data['subscribers'], name="Subscribers Histogram"),  # Plot subscribers as a histogram
    row=1, col=1  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the histogram
fig.update_xaxes(title_text="Subscribers", row=1, col=1)
fig.update_yaxes(title_text="Frequency", row=1, col=1)

# Add a Line Plot to the second subplot (row 1, col 2)
fig.add_trace(
    go.Scatter(x=data['rank'], y=data['video views'], mode='lines+markers', name="Video Views vs Rank"),  # Line plot with markers
    row=1, col=2  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the line plot
fig.update_xaxes(title_text="Rank", row=1, col=2)
fig.update_yaxes(title_text="Video Views", row=1, col=2)

# Prepare data for the Bar Chart: Calculate average engagement rate by category
engagement_data = data.groupby('category')['engagement_rate'].mean().sort_values(ascending=False)

# Add a Bar Chart to the third subplot (row 2, col 1)
fig.add_trace(
    go.Bar(x=engagement_data.index, y=engagement_data.values, name="Engagement Rate by Category"),  # Bar chart with categories and average engagement rates
    row=2, col=1  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the bar chart
fig.update_xaxes(title_text="Category", row=2, col=1)
fig.update_yaxes(title_text="Engagement Rate", row=2, col=1)

# Add a Scatter Plot to the fourth subplot (row 2, col 2)
fig.add_trace(
    go.Scatter(x=data['subscribers'], y=data['video views'], mode='markers',  # Scatter plot with subscribers and video views
               name="Subscribers vs Video Views"),
    row=2, col=2  # Specify the location in the subplot grid
)
# Set x-axis and y-axis labels for the scatter plot
fig.update_xaxes(title_text="Subscribers", row=2, col=2)
fig.update_yaxes(title_text="Video Views", row=2, col=2)

# Update the overall layout of the figure
fig.update_layout(
    height=800,  # Set the height of the entire figure
    width=1500,  # Set the width of the entire figure
    title_text="Subplot with 4 Different Graphs",  # Title of the figure
    showlegend=True  # Enable the legend to show plot labels
)

# The plot can now be displayed using fig.show() or saved as a
fig.show()
```

```python
# Aggregate data by category and country to sum video views
category_country_views = data.groupby(['category', 'Country'])['video views'].sum().reset_index()

# Create the initial bar chart
fig = px.bar(
    category_country_views,
    x='category',
    y='video views',
    color='category',
    title="Total Video Views by Category",
    labels={'video views': 'Video Views', 'category': 'Category'}
)

# Add dropdown for countries (multi-select logic implemented manually)
unique_countries = category_country_views['Country'].unique()

# Create a dropdown for country selection
fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                dict(label="All Countries",
                     method="update",
                     args=[{"visible": [True] * len(category_country_views)},
                           {"title": "Total Video Views by Category (All Countries)"}]),
                *[
                    dict(label=country,
                         method="update",
                         args=[{"visible": category_country_views['Country'] == country},
                               {"title": f"Total Video Views by Category ({country})"}])
                    for country in unique_countries
                ],
                dict(label="Clear",
                     method="update",
                     args=[{"visible": [False] * len(category_country_views)},
                           {"title": "Total Video Views by Category (No Country Selected)"}])
            ],
            direction="down",
            showactive=True,
            x=0.1,
            y=1.15,
        )
    ]
)

# Show the chart
fig.show()
```
